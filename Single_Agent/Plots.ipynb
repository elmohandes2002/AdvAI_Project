```python
import pandas as pd
import matplotlib.pyplot as plt
import re
from matplotlib.ticker import MaxNLocator  # <-- Add this import

# Load Excel file
df = pd.read_excel("/content/2000_300.xlsx")
logs = df.iloc[:, 0]

# Parse using regex
episodes = []
rewards = []
tiles_cleaned = []

for line in logs:
    match = re.search(r"Episode (\d+) finished\. Total reward: ([\-\d.]+), Cleaned: (\d+) tiles", str(line))
    if match:
        episodes.append(int(match.group(1)))
        rewards.append(float(match.group(2)))
        tiles_cleaned.append(int(match.group(3)))

# Plotting rewards
plt.figure(figsize=(10, 5))
plt.plot(episodes, rewards, label="Reward", color='blue')
plt.xlabel("Episode")
plt.ylabel("Reward")
plt.title("Episode vs Reward")
plt.grid(True)
plt.legend()
plt.show()

# Plotting tiles cleaned with whole number y-axis
plt.figure(figsize=(10, 5))
plt.plot(episodes, tiles_cleaned, label="Tiles Cleaned", color='green')
plt.xlabel("Episode")
plt.ylabel("Tiles Cleaned")
plt.title("Episode vs Tiles Cleaned")
plt.grid(True)
plt.legend()
plt.gca().yaxis.set_major_locator(MaxNLocator(integer=True))  # <-- Force whole number ticks
plt.show()
```
